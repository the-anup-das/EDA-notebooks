```python
# Loading data
salaries = pd.read_csv('../input/data-science-job-salaries/ds_salaries.csv')
```

```python
salaries.head()
```

```python
# Function query to execute SQL queries
def query(query):
    return ps.sqldf(query)
```

```python
# Showing Top 5 rows of data
query("""SELECT * FROM salaries LIMIT 5
""")
```

<a id = "5"></a>
<h1 style="padding:10px;background-color:#3498db;opacity: 0.8;margin:0;color:white;font-family:newtimeroman;font-size:200%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Data Preprocessing ‚öíÔ∏è</h1>

```python
salaries = salaries.drop('Unnamed: 0', axis = 1)
```

```python
salaries.isna().sum()
```

```python
salaries["employee_residence"]
```

```python
# Converting countries code to country names
salaries["employee_residence"] = coco.convert(names=salaries["employee_residence"], to="name")
salaries["company_location"] = coco.convert(names=salaries["company_location"], to="name")
```

```python
salaries["employee_residence"]
```

```python
# Replacing values in column - experience_level :
salaries['experience_level'] = query("""SELECT 
                                          REPLACE(
                                            REPLACE(
                                              REPLACE(
                                                REPLACE(
                                                  experience_level, 'MI', 'Mid level'), 
                                                                    'SE', 'Senior Level'), 
                                                                    'EN', 'Entry Level'), 
                                                                    'EX', 'Expert Level') 
                                        FROM 
                                          salaries""")
```

```python
salaries['experience_level']
```

```python
# Replacing values in column - experience_level :
salaries['employment_type'] = query("""SELECT 
                                          REPLACE(
                                            REPLACE(
                                              REPLACE(
                                                REPLACE(
                                                  employment_type, 'PT', 'Part Time'), 
                                                                    'FT', 'Full Time'), 
                                                                    'FL', 'Freelance'), 
                                                                    'CT', 'Contract') 
                                        FROM 
                                          salaries""")
```

```python
salaries['employment_type']
```

```python
# Replacing values in column - company_size :
salaries['company_size'] = query("""SELECT 
                                       REPLACE(
                                         REPLACE(
                                           REPLACE(
                                             company_size, 'M', 'Medium'), 
                                                           'L', 'Large'), 
                                                           'S', 'Small') 
                                    FROM 
                                       salaries""")
```

```python
salaries['overseas'] = [False if r[1].employee_residence==r[1].company_location else True for r in salaries.iterrows()]
```

```python
salaries['company_size']
```

```python
# Replacing values in column - remote_ratio :
salaries['remote_ratio'] = query("""SELECT 
                                        REPLACE(
                                          REPLACE(
                                            REPLACE(
                                              remote_ratio, '100', 'Fully Remote'), 
                                                            '50', 'Partially Remote'), 
                                                            '0', 'Non Remote Work') 
                                    FROM 
                                      salaries""") 
```

```python
salaries.head()
```

<div class="alert alert-block alert-info">
In <b>preprocessing step</b> first I converted country ISO code to country full name and replace many abbreviation to unit using <b>SQL</b> queries.
</div>

```python
# use country converter (coco.convert) to convert the country name to ISO3 standard
salaries['employee_residence_short'] = coco.convert(names=salaries.employee_residence, to='ISO3')
salaries['company_location_short'] = coco.convert(names=salaries.company_location, to='ISO3')
```

The sum of salaries might not be a correct measure for comparison as the entries in a particular country might be more than other. So, we plot the mean keeping the log10 scale. This gives a much better idea of salaries across the world.

```python
df = salaries
df["company_country"] = coco.convert(names = salaries["company_location"], to = 'name_short')

temp_df = df.groupby('company_country')['salary_in_usd'].sum().reset_index()
temp_df['salary_scale'] = np.log10(df['salary_in_usd'])

fig = px.choropleth(temp_df, locationmode = 'country names', locations = "company_country",
                   color = "salary_scale", hover_name = "company_country",
                   hover_data = temp_df[['salary_in_usd']], 
                    color_continuous_scale = 'Jet',
                   )
fig.update_layout(title={'text':'Salaries across the World', 
                         'xanchor': 'center','x':0.5})
fig.show()
```

```python
emp_per_country = salaries.groupby('employee_residence')['work_year'].count()
emp_per_country_df = pd.DataFrame(emp_per_country)
emp_per_country_df.columns = ['count']
emp_per_country_df.reset_index(inplace=True)
fig  = px.choropleth(emp_per_country_df,locationmode = 'country names',
                     locations = 'employee_residence',
                    color = 'count',
                    hover_name = 'employee_residence',
                    color_continuous_scale = 'Jet')
fig.update_layout(title={'text':'Based on Employee Residence', 
                         'xanchor': 'center','x':0.5})
fig.show()
```

```python
emp_per_country = salaries.groupby('company_location_short')['work_year'].count()
emp_per_country_df = pd.DataFrame(emp_per_country)
emp_per_country_df.columns = ['count']
emp_per_country_df.reset_index(inplace=True)
fig  = px.choropleth(emp_per_country_df, locations = 'company_location_short',
                    color = 'count',
                    hover_name = 'company_location_short',
                    color_continuous_scale = 'Jet')
fig.update_layout(title={'text':'Based on Company Location', 'xanchor': 'center','x':0.5})
fig.show()
```
